Creating views in a database can simplify complex queries, abstract underlying table structures, and provide a more user-friendly interface for data retrieval. Based on our social media marketing agency database schema, we are going to create two views, one that showcases the top-performing posts across campaigns, based on engagement or other metrics. This view can guide content strategy by highlighting successful post types.

```python
%%sql
CREATE VIEW TopPerformingPosts AS
SELECT
    p.CampaignID,
    p.PostID,
    p.Content,
    a.Engagement,
    a.Conversion
FROM Post p
JOIN Analytics a ON p.PostID = a.PostID
ORDER BY a.Engagement DESC
LIMIT 10;
```

The other view we will be creating shows engagement metrics (likes, shares, comments) for each platform, helping us identify which platforms are most effective for engagement.

```python
%%sql
CREATE VIEW PlatformEngagement AS
SELECT
    p.Name AS PlatformName,
    p.PlatformID,
    COALESCE(SUM(po.Likes), 0) AS TotalLikes,
    COALESCE(SUM(po.Shares), 0) AS TotalShares,
    COALESCE(SUM(po.Comments), 0) AS TotalComments
FROM Platform p
LEFT JOIN Post po ON p.PlatformID = po.PlatformID
LEFT JOIN Analytics a ON po.PostID = a.PostID
GROUP BY p.PlatformID;
```

Now it's time to create a funtion, creating functions in a database can help encapsulate complex logic, improve code reusability, and simplify the execution of common tasks. Based on our social media marketing agency database schema, lets build a function that calculates the average engagement (likes, shares, comments) for a given platform. This can help identify which platforms are consistently performing well.

```python
%%sql
CREATE or REPLACE FUNCTION EngagementRate(
    likes INT, 
    shares INT, 
    comments INT, 
    impressions INT
) 
RETURNS FLOAT
LANGUAGE plpgsql
AS $$
DECLARE
   TotalEngagement FLOAT; 
BEGIN
   TotalEngagement := likes + shares + comments;
   
   IF impressions > 0 THEN
       TotalEngagement := (TotalEngagement::FLOAT / impressions) * 100.0;
   ELSE
       TotalEngagement := 0.0;
   END IF;
   
   RETURN TotalEngagement;
END;
$$;
```

Lets call this funtion using a particular post with postID '1' to see its average engagement rate.

```python
%%sql
SELECT CONCAT(EngagementRate(150, 50, 30, 1000), '%') AS CalculatedEngagementRate;
```

Finally we create a create a stored procedure that archives old data from the Post and Analytics tables to separate archival tables. This can help manage database size and improve performance. First we create the table to house the archived data before creating the stored procedure.

```python
%%sql 
CREATE TABLE PostArchive(
      PostID INT, 
      CampaignID INT, 
      PlatformID INT, 
      Content TEXT, 
      PostDate DATE, 
      Likes INT, 
      Shares INT, 
      Comments INT,
      FOREIGN KEY (PostID) REFERENCES Post(PostID)
);
```

```python
%%sql
CREATE TABLE AnalyticsArchive(
     AnalyticsID INT, 
     PostID INT, 
     Impressions INT, 
     Clicks INT, 
     Engagement INT, 
     Conversion INT,
     FOREIGN KEY (PostID) REFERENCES Post(PostID),
     FOREIGN KEY (AnalyticsID) REFERENCES Analytics(AnalyticsID)
);
```

```python
%%sql
CREATE OR REPLACE PROCEDURE ArchiveOldData(
    archive_date DATE
)
LANGUAGE plpgsql
AS $$
BEGIN
    
    INSERT INTO PostArchive (PostID, CampaignID, PlatformID, Content, PostDate, Likes, Shares, Comments)
    SELECT PostID, CampaignID, PlatformID, Content, PostDate, Likes, Shares, Comments
    FROM Post
    WHERE PostDate < archive_date;

   
    INSERT INTO AnalyticsArchive (AnalyticsID, PostID, Impressions, Clicks, Engagement, Conversion)
    SELECT AnalyticsID, PostID, Impressions, Clicks, Engagement, Conversion
    FROM Analytics
    WHERE PostID IN (SELECT PostID FROM Post WHERE PostDate < archive_date);  
    
END;
$$;
```

Lets execute the procedure and perform the archiving of old data from the Post and Analytics tables earlier than 2023-06-01 to the archival tables.

```python
%%sql
CALL ArchiveOldData('2023-06-01');
```
